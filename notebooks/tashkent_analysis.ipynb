```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

df = pd.read_excel('../data/uybor.xlsx')

avg_price_district = df.groupby('district')['price'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x=avg_price_district.index, y=avg_price_district.values, hue=avg_price_district.index, palette='magma', legend=False)
plt.title('Tumanlar bo\'yicha uylarning o\'rtacha narxi (USD)', fontsize=15)
plt.ylabel('O\'rtacha narx')
plt.xticks(rotation=45)

plt.savefig('avg_price_district.png', dpi=300, bbox_inches='tight')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='size', y='price', alpha=0.6, color='darkblue')
plt.title('Uy maydoni va narxlari o\'rtasidagi bog\'liqlik', fontsize=15)
plt.xlabel('Maydon (m2)')
plt.ylabel('Narx (USD)')
plt.grid(True, linestyle='--', alpha=0.5)

plt.savefig('price_scatter.png', dpi=300, bbox_inches='tight')
plt.show()

numeric_df = df.select_dtypes(include=['float64', 'int64'])
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Ustunlar o\'rtasidagi bog\'liqlik darajasi (Heatmap)', fontsize=15)

plt.savefig('heatmap.png', dpi=300, bbox_inches='tight')
plt.show()
```
